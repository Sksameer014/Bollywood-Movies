Geting features selection

```python
df_fea=df[['movie_id', 'year', 'genre', 'overview', 'director',
       'cast']].fillna(' ')
```

```python
df_fea.shape
```

```python
df_fea
```

```python
x=df_fea['genre']+ ' ' +df_fea['overview']+ ' ' +df_fea['director']+ ' ' +df_fea['cast']
```

```python
x
```

```python
x.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf=TfidfVectorizer()
x=tf.fit_transform(x)
```

```python
x.shape
```

```python
print(x)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sml_scr=cosine_similarity(x)
```

```python
sml_scr
```

```python
sml_scr.shape
```

```python
FMN=input("Enter Your Favourite Movie Name: ")
```

```python
AMTitle=df['movie_name'].tolist()
```

```python
import difflib as dl
```

```python
MRec=dl.get_close_matches(FMN, AMTitle)
print(MRec)
```

```python
Close_Match=MRec[0]
print(Close_Match)
```

```python
Index_CMM=df[df.movie_name == Close_Match]['movie_id'].values[0]
print(Index_CMM)
```

```python
Rec_scr=list(enumerate(sml_scr[Index_CMM]))
print(Rec_scr)
```

```python
len(Rec_scr)
```

```python
sort_smlM=sorted(Rec_scr, key=lambda x:x[1], reverse=True)
```

```python
print(sort_smlM)
```

```python
print("Top 30 Movies Suggested for You: \n")
i=1
for movie in sort_smlM:
    index=movie[0]
    tfindex=df[df.index==index]['movie_name'].values[0]
    if(i<31):
        print(i,'-',tfindex)
        i+=1
```

Top 10 Movie Recommendation 

```python
MN=input("Enter Your favourite movie name: ")
LAtitles=df['movie_name'].tolist()
FCM=dl.get_close_matches(MN,LAtitles)
CM=FCM[0]
indexM=df[df.movie_name==CM]['movie_id'].values[0]
rec_scr=list(enumerate(sml_scr[indexM]))
sortsmlM=sorted(rec_scr, key=lambda x:x[1], reverse=True)
print("Top 10 Movies Suggested for you: \n")
i=1
for movie in sortsmlM:
    index=movie[0]
    TitleFI=df[df.movie_id==index]['movie_name'].values
    if (i<11):
        print(i,'-',TitleFI)
        i+=1
```
